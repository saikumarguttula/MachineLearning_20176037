### Importing all the required libraries

```python
import matplotlib.pyplot as plt
import os,cv2
import pandas as pd
import numpy as np
```

### Loading Data

```python
from PIL import Image
def load_images(folder):
    image_list = []
    for filename in os.listdir(folder):
        image = Image.open(os.path.join(folder,filename))
        image = np.array(image.convert('L')) 
        image = image / 255 
        print(image.shape)
        print(image)
        if image is not None:
            image_list.append(image)
    return image_list
```

```python
data=[]
FolderName = "att_faces"
for i in range(1,41):
    for j in load_images(os.path.join(FolderName,"s"+str(i))):
        data.append(list((j,i)))
```

### Adding Label Column

```python
df = pd.DataFrame(data,columns=['Image','Label'])
```

### Reshaping the data in matrix form to a single array

```python
image_Count=len(df['Image'])
for i in range(image_Count):
    img = df['Image'][i]
    df['Image'][i] = img.reshape(img.shape[0]*img.shape[1])

# print(df['Image'][1])
```

### Splitting of Train & Test data

```python
train_image = []
test_image = []
train_label = [] 
test_label= []
```

```python
# Selecting an image from each folder for test data
muls=[10*i for i in range(1,41) ]
for i in range(0,400):
    if(i>0  and (i+1 in  muls)):
        test_image.append(df['Image'][i])
        test_label.append(df['Label'][i])
    else:
        train_image.append(df['Image'][i])
        train_label.append(df['Label'][i])
```

```python
train_image_vectors = []
for i in train_image:
    train_image_vectors.append(i.tolist())
test_image_vectors = []
for j in test_image:
    test_image_vectors.append(j.tolist())
```

```python
for i in range(360):
    for j in range(10304):
        train_image_vectors[i][j]=np.asscalar(np.asarray(train_image_vectors[i][j]))
for i in range(40):
    for j in range(10304):
        test_image_vectors[i][j]=np.asscalar(np.asarray(test_image_vectors[i][j]))
```

### Converting images to matrix

```python
images_Matrix = np.matrix(train_image_vectors)
```

### PCA with 95% of variance explained

```python
from sklearn import decomposition
pca = decomposition.PCA(.95)
pca.fit(images_Matrix)
```

```python
pca.n_components_
```

* 176 principal components are enough to explain 95% of the variance

### Applying PCA on train & test data

```python
train_image = pca.transform(np.asmatrix(train_image_vectors,dtype='float64'))
test_image = pca.transform(np.asmatrix(test_image_vectors,dtype='float64'))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression(solver = 'lbfgs',multi_class='multinomial')
logisticRegr.fit(train_image, train_label)
```
