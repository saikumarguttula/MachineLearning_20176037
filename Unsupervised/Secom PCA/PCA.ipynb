### Importing all the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("./Data/secom.data",delimiter=" ",header=None)
```

### Exploration of data

```python
data.head()
```

```python
data.describe()
```

### Missing value imputation

```python
data_after_imputation = data.fillna(data.median())
```

```python
data_after_imputation.describe()
```

### Removing zero variance columns 

```python
cols = data.nunique()==1
```

```python
sampled_data=data_after_imputation.drop(list(data.columns[cols]),axis=1)
```

```python
sampled_data.shape
```

### Data Scaling

```python
from sklearn.preprocessing import StandardScaler
scaled_data=StandardScaler().fit_transform(sampled_data)
scaled_data.shape
scaled_data
```

```python
covariant_matrix = np.matmul(scaled_data.T,scaled_data)
```

```python
covariant_matrix.shape
```

### Eigen values & vectors

```python
from scipy.linalg import eigh
```

```python
values,vectors = eigh(covariant_matrix)
```

```python
vectors.shape
```

```python
print(values)
```

### PCA

```python
from sklearn import decomposition
pca = decomposition.PCA(scaled_data)
```

```python
pca.n_components = 474
pca_data = pca.fit_transform(scaled_data)
```

```python
pca.explained_variance_
```

```python
variance_of_each_pc = pca.explained_variance_ratio_
variance_of_each_pc
```

```python
cumulativeVariance = pca.explained_variance_ratio_.cumsum()
```

```python
plt.figure(1, figsize=(6, 4))
plt.plot(cumulativeVariance, linewidth=2)
plt.xlabel('n_components')
plt.ylabel('cumulativeVariance')
plt.grid(True)
plt.show()
```

* From the above plot, It is clear that about 200 - 210 PC's explains 98% of the total variance 

```python
pca.n_components = 7
seven_principal_Components = pca.fit_transform(scaled_data)
principal_DF = pd.DataFrame(data = seven_principal_Components,columns = ['Pc1','Pc2','Pc3','Pc4','Pc5','Pc6','Pc7'])
```
